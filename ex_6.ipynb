```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing import sequence
from sklearn.model_selection import train_test_split
from keras import layers
from keras.models import Model
```

```python

data = pd.read_csv("ner_dataset.csv", encoding="latin1")
```

```python


data.head(50)
```

```python

data = data.fillna(method="ffill")
```

```python

data.head(50)
```

```python

print("Unique words in corpus:", data['Word'].nunique())
print("Unique tags in corpus:", data['Tag'].nunique())
```

```python
words=list(data['Word'].unique())
words.append("ENDPAD")
tags=list(data['Tag'].unique())
```

```python

print("Unique tags are:", tags)
```

```python

num_words = len(words)
num_tags = len(tags)
```

```python

num_words
```

```python

class SentenceGetter(object):
    def __init__(self, data):
        self.n_sent = 1
        self.data = data
        self.empty = False
        agg_func = lambda s: [(w, p, t) for w, p, t in zip(s["Word"].values.tolist(),
                                                           s["POS"].values.tolist(),
                                                           s["Tag"].values.tolist())]
        self.grouped = self.data.groupby("Sentence #").apply(agg_func)
        self.sentences = [s for s in self.grouped]
```

```python
 def get_next(self):
        try:
            s = self.grouped["Sentence: {}".format(self.n_sent)]
            self.n_sent += 1
            return s
        except:
            return None
```

```python

getter = SentenceGetter(data)
sentences = getter.sentences
```

```python

len(sentences)
```

```python


sentences[0]
```

```python

word2idx = {w: i + 1 for i, w in enumerate(words)}
tag2idx = {t: i for i, t in enumerate(tags)}
```

```python

word2idx
```

```python

plt.hist([len(s) for s in sentences], bins=50)
plt.show()
```

```python


X1 = [[word2idx[w[0]] for w in s] for s in sentences]
```

```python

type(X1[0])
```

```python

X1[0]
```

```python

max_len = 50
```

```python

X = sequence.pad_sequences(maxlen=max_len,
                  sequences=X1, padding="post",
                  value=num_words-1)
```

```python

X[0]
```

```python

y1 = [[tag2idx[w[2]] for w in s] for s in sentences]
```

```python

y = sequence.pad_sequences(maxlen=max_len,
                  sequences=y1,
                  padding="post",
                  value=tag2idx["O"])
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2, random_state=1)
```

```python

X_train[0]
```

```python

y_train[0]
```

```python
input_word = layers.Input(shape=(max_len,))
# Write your code here
output = layers.Dense(num_tags)(input_word) # Example dense layer - replace with your desired layers
model = Model(input_word, output)
```

```python

print('Name:  THAMARAISELVAN V               Register Number:  212221230115     ')
model.summary()
```

```python
model.compile(optimizer="adam",
              loss="categorical_crossentropy",
              metrics=y["accuracy"])
```

```python
input_word = layers.Input(shape=(max_len,))
# Ensure num_tags matches the number of unique tags (50 in this case)
output = layers.Dense(50)(input_word) # Output layer with the correct number of tags
model = Model(input_word, output)
```

```python
model.compile(optimizer="adam",
              loss="categorical_crossentropy",
              metrics=["accuracy"])
```

```python

history = model.fit(
    x=X_train,
    y=y_train,
    validation_data=(X_test,y_test),
    batch_size=32,
    epochs=3,
)
```

```python

metrics = pd.DataFrame(model.history.history)
metrics.head()
```

```python

print('Name:  THAMARAISELVAN V               Register Number: 212221230115      ')
metrics[['accuracy','val_accuracy']].plot()
```

```python

print('Name:   THAMARAISELVAN V              Register Number:  212221230115     ')
metrics[['loss','val_loss']].plot()
```
